## 1. Számok a táblán
Egy táblára felírtuk a $1,2,3,\ldots,1978$ számokat. A következőt ismételjük: <br>
kiválasztunk $2$ számot, letöröljük őket és felírjuk helyettük a *különbségüket*. <br>
Mutassuk meg, hogy a az utolsó szám *páratlan*.

### S:
invariáns keresése.

### P:
Írjunk programot, mely ezt a játékot szimulálja. Milyen nagy számra számíthatunk a végén?

[megoldások](#SZT_M)

<a id='SZT_PM'> Számok a táblán </a>


```julia
# függvényesen általában gyorsabb kódot eredményez
function szt(n)
    sz=collect(1:n)
    while n>1
        i=rand(1:n)
        if i!=n sz[i],sz[n]=sz[n],sz[i] end
        j=rand(1:n-1)
        if j!=(n-1) sz[j],sz[n-1]=sz[n-1],sz[j] end
        sz[n-1]=abs(sz[n-1]-sz[n])
        n-=1
    end
    sz[1]
end

# az állítás random tesztelése
for ism=1:1000
    szt(1978)%2==1 || error("hiba")
end


# átlagos utolsó számok:
A=1000
B=2000
ISM=300
res=zeros(B)
for n in A:B
    s=0.0
    for ism in 1:ISM
        s+=szt(n)
    end    
    res[n]=(s/ISM)
end
using Plots
# elég szép egyenes
plot(A:B,res[A:B],label="")
# vagy mégsem:
plot!([A,B],[A/5.0,B/5.0],label="5y=x")
```
